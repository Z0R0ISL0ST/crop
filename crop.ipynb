```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import pickle

# Load dataset
csv_file = "cpdata.csv"
df = pd.read_csv(csv_file)

# Data preprocessing
df.drop('rainfall', axis=1, inplace=True)
X = df.drop('labels', axis=1)
y = df['labels']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree model
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
with open('crop_pred_dtree.pkl', 'wb') as f:
    pickle.dump(dtree, f, protocol=4)
print("Decision Tree Accuracy:", accuracy_score(y_test, dtree.predict(X_test)))

# Train Random Forest model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
with open('crop_pred_rand.pkl', 'wb') as f:
    pickle.dump(rf, f, protocol=4)
print("Random Forest Accuracy:", accuracy_score(y_test, rf.predict(X_test)))

# Train KNN model
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(X_train, y_train)
with open('crop_pred_knn.pkl', 'wb') as f:
    pickle.dump(knn, f, protocol=4)
print("KNN Accuracy:", accuracy_score(y_test, knn.predict(X_test)))

# Train SVM model
svm = SVC()
svm.fit(X_train, y_train)
with open('crop_pred_svc.pkl', 'wb') as f:
    pickle.dump(svm, f, protocol=4)
print("SVM Accuracy:", accuracy_score(y_test, svm.predict(X_test)))
```
